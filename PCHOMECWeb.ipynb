## **POHOME大類別24H**


```python
import requests
from bs4 import BeautifulSoup as bs

url = 'https://24h.pchome.com.tw/'
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/52.0.2743.116 Safari/537.36'}
res = requests.get(url, headers=headers).text
soup = bs(res, 'lxml')
#links = soup.find('dl', class_ = 'site_roadsign hp').prettify()
links = soup.find_all('a', class_ = 'sign')
data=[]
a = 0
for link in links:
    if a < 12:
        data.append(str(links[a]).replace('</a>','',1).lstrip('<')[-4::].lstrip('">'))
        a+=1

print(data)
```
